### 3)  Apply Scaling to AGE attribute with min max, decimal scaling and z score.


```python
dataFrame.fillna(dataFrame.Age.mean() , inplace = True)
data1 = dataFrame.copy()
min_value = data1['Age'].min()
max_value = data1['Age'].max()

data1['min_max_scaling'] = (data1['Age'] - min_value)/(max_value - min_value)
data1
```

```python
# dataFrame.fillna(dataFrame.Age.mean() , inplace = True)
# data1 = dataFrame.copy()
max_value = data1['Age'].max()
digits = len(str(int(max_value)))

# print(digits)

data1['decimal_Scaling'] = (data1['Age'] ) / (10** digits)
data1
```

```python
mean_value = data1.Age.mean()
strd_value = data1.Age.std()

data1['z_score'] = (data1['Age'] - mean_value)/strd_value
data1
```
