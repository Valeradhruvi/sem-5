# Step 1: Load Data

```python
import pandas as pd
```

```python
df=pd.read_csv("OnlineRetail.csv",encoding='ISO-8859-1')
df
```

```python
df.isnull().sum()
```

# Step 2: Drop Missing Invoices or Descriptions

```python
df.dropna(subset=['InvoiceNo', 'Description'], inplace=True)

```

```python
df.isnull().sum()
```

# Step 3: Remove Canceled Invoices <BR> Start with letter 'c'

```python
df.drop(df[df['InvoiceNo'].str.startswith('C')].index, inplace=True)
```

```python
tocheck = df[df['InvoiceNo'].str.startswith('C')]
print(tocheck)
```

# Step 4: Keep Only Positive Quantities

```python
df = df[df['Quantity'] > 0].copy()
df
```

# Step 5: Standardize Item Descriptions | Lower case and remove space

```python
df['Description'] = df['Description'].str.lower().str.strip()

df
```

# Step 6: Filter for a Single Country (e.g., United Kingdom)

```python
uk = df[df['Country'] == 'United Kingdom']
uk
```

# Step 7: Remove Duplicates

```python
dup=df.drop_duplicates(inplace=True)
print(dup)
```

# Step 8: Add TotalPrice column , TotalPrice = Quantity *  UnitPrice

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
df
```

# Step 9 : Check for Top 10 Selling Products

```python
topproducts = df.groupby('Description')['Quantity'].sum().sort_values().tail(10)
print(topproducts)
```

# Step 10 : Identify Most Active Customers

```python
actuser = df.groupby('CustomerID')['InvoiceNo'].sum().sort_values().tail(10)
print(actuser)
```
